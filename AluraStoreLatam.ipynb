### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

## tienda.head() # muestra las 5 primeras filas (verificar datos)
```

#1. Análisis de facturación



```python
# Calcula la suma de todos los valores en la columna 'Precio' para cada tienda
total_tienda1 = tienda1['Precio'].sum()
total_tienda2 = tienda2['Precio'].sum()
total_tienda3 = tienda3['Precio'].sum()
total_tienda4 = tienda4['Precio'].sum()

# Muestra en pantalla el total facturado de cada Tienda
print(f"El valor Total facturado de la Tienda 1: $ {total_tienda1}")
print(f"El valor Total facturado de la Tienda 2: $ {total_tienda2}")
print(f"El valor Total facturado de la Tienda 3: $ {total_tienda3}")
print(f"El valor Total facturado de la Tienda 4: $ {total_tienda4}")
```

# 2. Ventas por categoría

```python
# Cuenta cuántos productos hay por cada categoría en la tienda1 y guarda el resultado en vpc_tienda1
vpc_tienda1 = tienda1['Categoría del Producto'].value_counts()
vpc_tienda2 = tienda2['Categoría del Producto'].value_counts()
vpc_tienda3 = tienda3['Categoría del Producto'].value_counts()
vpc_tienda4 = tienda4['Categoría del Producto'].value_counts()

# Imprime en pantalla el conteo de productos por categoría en la Tienda 1 (resultado de vpc_tienda1)
print(f'Tienda 1: {vpc_tienda1}')
print('-'*35)
print(f'Tienda 2: {vpc_tienda2}')
print('-'*35)
print(f'Tienda 3: {vpc_tienda3}')
print('-'*35)
print(f'Tienda 4: {vpc_tienda4}')
print('~'*35)
```

```python
# Agrupa los productos de tienda1 por categoría, suma los precios de cada grupo y ordena las categorías de mayor a menor total
precio_categoria1 = tienda1.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
precio_categoria2 = tienda2.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
precio_categoria3 = tienda3.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)
precio_categoria4 = tienda4.groupby('Categoría del Producto')[['Precio']].sum().sort_values('Precio', ascending=False)

# Imprime en pantalla el total de precios agrupados por categoría de producto en la Tienda 1 (precio_categoria1)
print (f'Tienda 1:{precio_categoria1}')
print('-'*35)
print (f'Tienda 2:{precio_categoria2}')
print('-'*35)
print (f'Tienda 3:{precio_categoria3}')
print('-'*35)
print (f'Tienda 4:{precio_categoria4}')
```

# 3. Calificación promedio de la tienda


```python
# Lista de tiendas
tiendas = [tienda1, tienda2, tienda3, tienda4]

# Calcular la valoración media por tienda
for i, t in enumerate(tiendas, start=1):
    valoracion_media = t["Calificación"].mean()
    print(f"Valoración media - Tienda {i}: {valoracion_media:.2f}")
```

# 4. Productos más y menos vendidos

```python
def resumen_ganancias_por_tienda(tienda, numero_tienda):
    # Calcular monto total por producto
    ganancias_producto = tienda.groupby("Producto")["Precio"].sum()

    # Top 3 productos con más ganancias
    top_mas_ganancias = ganancias_producto.sort_values(ascending=False).head(3)

    # Top 3 productos con menos ganancias
    top_menos_ganancias = ganancias_producto.sort_values(ascending=True).head(3)

    # Mostrar resultados
    print(f"--- Tienda {numero_tienda} ---")
    print("Top 3 productos con más ganancias:")
    for producto, monto in top_mas_ganancias.items():
        print(f"{producto}: ${monto}")

    print("\nTop 3 productos con menos ganancias:")
    for producto, monto in top_menos_ganancias.items():
        print(f"{producto}: ${monto}")

    print("-" * 50)


# Aplicar la función a todas las tiendas
for i, t in enumerate(tiendas, start=1):
    resumen_ganancias_por_tienda(t, i)
```

# 5. Envío promedio por tienda

```python
# Lista de tiendas
tiendas = [tienda1, tienda2, tienda3, tienda4]

# Calcular el valor promedio del envío por tienda
for i, t in enumerate(tiendas, start=1):
    envio_promedio = t["Costo de envío"].mean()
    print(f"Valor del envío promedio - Tienda {i}: {envio_promedio:.2f}")
```

```python
def envio_promedio_por_tienda(tienda, numero_tienda):
    # Calcular el valor promedio del envío
    promedio_envio = tienda["Costo de envío"].mean()
    print(f"Tienda {numero_tienda} - Valor de envío promedio: ${promedio_envio:.2f}")

for i, t in enumerate(tiendas, start=1):
    envio_promedio_por_tienda(t, i)
```

# Preparacion para comenzar a graficar

```python
import matplotlib.pyplot as plt

# Calcular ingresos totales por tienda
ingresos_tiendas = []
for i, t in enumerate(tiendas, start=1):
    ingresos_totales = t["Precio"].sum()
    ingresos_tiendas.append((f"Tienda {i}", ingresos_totales))

# Convertir a listas separadas para graficar
nombres_tiendas = [tienda for tienda, _ in ingresos_tiendas]
valores_ingresos = [ingreso for _, ingreso in ingresos_tiendas]

# Definir colores: rojo para el menor, verde para los demás
min_ingreso = min(valores_ingresos)
colores = ["red" if ingreso == min_ingreso else "green" for ingreso in valores_ingresos]

# Gráfico de barras con colores personalizados
plt.figure(figsize=(8,5))
plt.bar(nombres_tiendas, valores_ingresos, color=colores)
plt.title("Ingresos totales por tienda")
plt.ylabel("Monto total ($)")
plt.xlabel("Tiendas")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Valores de valoración media por tienda
valoraciones = [3.98, 4.04, 4.05, 4.00]
nombres_tiendas = [f"Tienda {i}" for i in range(1, len(valoraciones)+1)]

# Identificar mínimo y máximo
min_val = min(valoraciones)
max_val = max(valoraciones)

# Colores dinámicos
colores = ["red" if v == min_val else "blue" if v == max_val else "green" for v in valoraciones]

# Gráfico de puntos
plt.figure(figsize=(8,5))
plt.scatter(nombres_tiendas, valoraciones, color=colores, s=200, marker="o")
plt.title("Valoración media por tienda")
plt.ylabel("Calificación promedio")
plt.xlabel("Tiendas")
plt.ylim(1,5)  # escala de referencia
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Recorremos cada tienda
for i, t in enumerate(tiendas, start=1):
    # Contar cantidad de ventas por categoría
    ventas_categoria = t.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)

    # Gráfico de pastel
    plt.figure(figsize=(6,6))
    plt.pie(
        ventas_categoria,
        labels=ventas_categoria.index,   # nombres de categorías
        autopct="%1.1f%%",               # porcentaje con un decimal
        startangle=90
    )
    plt.title(f"Distribución de categorías más vendidas - Tienda {i}")
    plt.legend(title="Categorías",
               loc="center left",
               bbox_to_anchor=(1.3,0.5)
    )
    plt.show()
```
